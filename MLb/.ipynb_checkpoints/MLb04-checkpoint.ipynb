```python
#to store data of anyone particular column
#synatx: new_var=df.column_name OR new_var=df['column_name']
fname=df.fruit_name
fname
```

```python
#to find number of unique entries in a particular column
#m1: use df.column_name.nunique()
#can replace df.fruit_name=fname also
df.fruit_name.nunique()
```

```python
#m2: use len(df.column_name.unique())
len(fname.unique())
```

```python
#m3: assigning the unique elements to a variable and finding its size
fruitn=fname.unique()
fruitn
```

```python
fruitn.size
```

```python
#To find out the exact count of each and every fruit
#apple - 19
#mandarin - 5
#Ornages - 19
#lemons - 16 
#This count was done manually
#use var_name=df.groupby('column_name'/df.column_name).size()
#here .groupby sets the values in alphabetical order
dsize=df.groupby(fname).size()
dsize
```

```python
dsize.size
```
