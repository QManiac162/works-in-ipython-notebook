```python
#We want to consider only the numeric data
#So we will create a new dataframe with only numeric data
df1_numeric = df1.select_dtypes(include = ['float64','int64'])
df1_numeric
```

```python
#Now we have to remove or drop the CarID and Symboling column
#axis = 1 means column
#axis = 0 means row
df1_numeric = df1_numeric.drop(['symboling','car_ID'],axis = 1)
df1_numeric
```

```python
df1_numeric.info()
```

```python
x1=df1_numeric.iloc[:,0:13].values
y1=df1_numeric.iloc[:,13].values
print(x1,'\n',y1)
```

```python
#train and test variables
#from sklearn.model_selection= package import train_test_split= library
from sklearn.model_selection import train_test_split as tts
x1_train,x1_test,y1_train,y1_test=tts(x1,y1,random_state=0)
'''
The random state hyperparameter in the train_test_split() function controls the shuffling process. 
With random_state=None , we get different train and test sets across different executions and the 
shuffling process is out of control. With random_state=0 , we get the same train and test sets 
across different executions.
'''
#Whatever data splitting /data allocation happens to the 
#xtrain,x_test,ytrain,ytest variables , we want those allocated values to remain constant
#By default the training variables get 75 % and testing variables get 25%
```

```python
print(x1.shape,'\n',x1_train.shape,'\n',x1_test.shape)
```

```python
print(y1.shape,'\n',y1_train.shape,'\n',y1_test.shape)
```

```python
#SCALING or NORMALISATION -DONE ONLY FOR INPUTS
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x1_train = mms.fit_transform(x1_train)
x1_test = mms.fit_transform(x1_test)
```

```python
#MODEL FITTING
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x1_train,y1_train)
```

```python
#PREDICT THE OUTPUT
#By taking the input testing data, we predict the output
#The predicted output is as follows
y1_pred=lr.predict(x1_test)
y1_pred
```
