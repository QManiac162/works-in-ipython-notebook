```python
#5.BINARY IMAGE CONVERSION(HIGH CONTRAST)
import cv2
img = cv2.imread(r'C:/Users/DEEP/Machine Learning/images/copy.jpg',0)#0 converts image into grayscale
cv2.imshow('GRAY SCALE IMAGE',img)
cv2.waitKey(2000)

ret,binary = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
#ret,binary are the 2 variables to be taken .Ret is a dummy variable
#and Binary is the actual variable 

cv2.imshow('Binary',binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
'''
IMAGE THRESHOLDING:
thresholding means  An image processing method that creates a bitonal (aka binary) image based on setting 
a threshold value on the pixel intensity of the original image. While most commonly applied to grayscale 
images, it can also be applied to color images.

For every pixel, the same threshold value is applied. If the pixel value is smaller than the threshold, 
it is set to 0, otherwise it is set to a maximum value. The function cv.threshold is used to apply the 
thresholding. The first argument is the source image, which should be a grayscale image. The second 
argument is the threshold value which is used to classify the pixel values. The third argument is the 
maximum value which is assigned to pixel values exceeding the threshold. OpenCV provides different types 
of thresholding which is given by the fourth parameter of the function. Basic thresholding as described 
above is done by using the type cv.THRESH_BINARY. All simple thresholding types are:

cv.THRESH_BINARY
cv.THRESH_BINARY_INV
cv.THRESH_TRUNC
cv.THRESH_TOZERO
cv.THRESH_TOZERO_INV
'''
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img=cv.imread(r'C:/Users/DEEP/Machine Learning/images/copy.jpg',0)
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#6.SOLID BACKGROUND(WHITE or BLACK BACKGROUND)
#WHITE
import numpy as np
import cv2
img=np.ones((500,500,3))#500 is length , 500 is width in pixels,3 - depth
#Img variable has an image with white background
cv2.imshow('WHITE BACKGROUND',img)
cv2.waitKey(1000)

#BLACK
import numpy as np
import cv2
img=np.zeros((500,500,3))#500 is length , 500 is width in pixels,3 - depth
#Img variable has an image with white background
cv2.imshow('BLACK BACKGROUND',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#7.SOLID COLORS(RGB)
#RED COLOR
import numpy as np
import cv2
img = np.zeros((450,670,3)) # BLACK BACKGROUND
img[:] = 0,0,255 #Assigning the color (B,G,R)
cv2.imshow('RED',img)
cv2.waitKey(1000)

#GREEN COLOR
import numpy as np
import cv2
img = np.zeros((450,670,3)) # BLACK BACKGROUND
img[:] = 0,255,0 #Assigning the color (B,G,R)
cv2.imshow('GREEN',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#8.Let us create an image of 300x300 pixels
#and divide the image into 3 parts of 3 different colors

import cv2
import numpy as np
#1st create a black background
image=np.zeros((300,300,3))
#image[y-coord,x-coord]
image[0:100,200:300]=0,255,0#GREEN
image[100:300,100:200]=255,0,0#BLUE
image[200:300,200:300]=0,0,255#RED
cv2.imshow('COLORS',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#9.CHECKER BOARD
import numpy as np
import cv2
#creates a black background of 200x200 pixels
img = np.zeros((200,200,3))
#that pixel is colored white according to alternate grid position
#no pixel assigned leaves it to default/ background color i.e black
img[0:100,0:100] = 255,255,255
img[100:200,100:200] = 255,255,255

cv2.imshow('CHECKER BOARD',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
